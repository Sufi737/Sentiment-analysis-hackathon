```python
def clean_tweet(tweet):
    tweet=re.sub(r'((w{3,}\.|https?|bit\.ly|[^\s]+\.(co|com)).[^\s]*|{link})',' ',tweet)
    tweet=re.sub(r'([a-zA-Z])\1{2,}',r'\1',tweet)
    # Smile -- :), : ), :-), (:, ( :, (-:, :')
    tweet = re.sub(r'(:\s?\)|:-\)|\(\s?:|\(-:|:\'\))', ' happy ', tweet)
    # Laugh -- :D, : D, :-D, xD, x-D, XD, X-D
    tweet = re.sub(r'(:\s?D|:-D|x-?D|X-?D)', ' happy ', tweet)
    # Love -- <3, :*
    tweet = re.sub(r'(<3|:\*)', ' happy ', tweet)
    # Wink -- ;-), ;), ;-D, ;D, (;,  (-;
    tweet = re.sub(r'(;-?\)|;-?D|\(-?;)', ' happy ', tweet)
    # Sad -- :-(, : (, :(, ):, )-:
    tweet = re.sub(r'(:\s?\(|:-\(|\)\s?:|\)-:)', ' sad ', tweet)
    # Cry -- :,(, :'(, :"(
    tweet = re.sub(r'(:,\(|:\'\(|:"\()', ' sad ', tweet)
    tweet=re.sub(r'[a-zA-z0-9_.+-]*[\@]+[a-zA-Z0-9_.+-]+',' ',tweet)
    tweet=re.sub(r'(^RT|[\s]+(RT|rt)[\s]+)',r' ',tweet)
    tweet=re.sub(r'\$\w*',' ',tweet)
    tweet=re.sub(r'&\w*',' ',tweet)
    
    tweet=tweet.encode('utf-8')
    tweet=re.sub(rb'[^\x00-\x7f]',rb' ',tweet)
    return tweet.decode('utf-8')
```

```python
lemmatizer = WordNetLemmatizer()
tokenizer2 = TweetTokenizer()
```

```python
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
```

```python
 
def treatment_tweet(tweet):
    tweet= tweet.lower()
    p =re.compile(r'\b(' + '|'.join(short_word_dict.keys()) + r')\b')
    tweet = p.sub(lambda x: short_word_dict[x.group()],tweet)
    q =re.compile(r'\b(' + '|'.join(negation_dic.keys()) + r')\b')
    tweet = q.sub(lambda x: negation_dic[x.group()],tweet)
    tweet = re.sub(r'[\s]+', ' ', tweet ) #Remove additional white spaces
    tweet = ' '.join([lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in tokenizer2.tokenize(tweet)])
    tweet = tweet.strip(' ')
    return tweet
```

```python
data=pd.read_csv('preprocess.csv')
sentiments=pd.read_csv('sentiment.csv')
test=pd.read_csv('preprocess_test.csv')
test.replace(np.nan,'neutral',inplace=True)
```

```python
vectorizer= TfidfVectorizer(lowercase=False,ngram_range=(1,2),max_df=1.0)
x_train_tfidf = vectorizer.fit_transform(data['tweet'])
```

```python
svc=SVC(C=100,gamma=0.01,kernel='rbf')
svc_model=svc.fit(x_train_tfidf,sentiments['sentiment'])
```

```python
tokenizer = Tokenizer()

tokenizer.fit_on_texts(data['tweet'])
total_words_train= len(tokenizer.word_index) + 1
total_words_test= len(tokenizer.word_index) + 1
```

```python
input_sequences=tokenizer.texts_to_sequences(data['tweet'].values)
test_sequence=tokenizer.texts_to_sequences(test['tweet'].values)
```

```python
max_sequence_len=max([len(x) for x in input_sequences])

train_data_X=np.array(pad_sequences(input_sequences,padding='post',maxlen=max_sequence_len))
test_data=np.array(pad_sequences(test_sequence,padding='post',maxlen=max_sequence_len))
```

```python
train_data_y=tf.keras.utils.to_categorical(sentiments['sentiment'],num_classes=4)
```

```python
def get_f1(y_true, y_pred): #taken from old keras source code
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val
```

```python
model = Sequential()
model.add(Embedding(total_words_train, 200, input_length=train_data_X.shape[1]))
model.add(Bidirectional(LSTM(150, return_sequences = True)))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words_train/2, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(128, activation='relu'))

model.add(Dense(4, activation='softmax'))

# Compile the network :
model.compile(loss = 'categorical_crossentropy', optimizer='adam',metrics=['acc',get_f1])
print(model.summary())
```

```python
history = model.fit(train_data_X,train_data_y, epochs=100,verbose=1)
```

```python
Y_pred = model.predict_classes(test_data)
```
